```python
import pandas as pd
import os
import glob
from pathlib import Path
from tqdm.auto import tqdm
```

```python
list_md = list(glob.glob(os.path.join(".","docs","*.md")))
```

```python
list_md[0]
```

```python
df=pd.read_json("data.jsonl",lines=True)
```

```python
df
```

```python
metadata={}
for i,v in df.iterrows():
    _d={
        "url":v["url"],
        "title":v["title"]
    }
    metadata[v["id"]]=_d
```

```python
metadata[44]
```

```python
import re
def remove_img_tags(data):
    p = re.compile(r'<img.*?/>')
    return p.sub('', data)
```

```python
list_data = []
k=1
for i in tqdm(list_md):
   try:
      id=int(Path(i).stem.split("-")[0])
      data = {"id":k,"metadata":metadata[id],"source":"openbase.in.th"}
      data["metadata"]["domain"]="literature"
      with open(i,"r") as f:
         data["text"]=remove_img_tags(f.read()).strip()
      list_data.append(data)
      k+=1
   except:
      continue
```
