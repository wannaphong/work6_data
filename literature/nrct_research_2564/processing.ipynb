```python
def get_text(name_th="-",name_en="-",keyword_th="-",keyword_en="-",objective="-",abstract_th="-",abstract_en="-",oecd1="-",oecd2="-"):
    text=f"""# ชื่อโครงการ
{name_th}

# ชื่อโครงการภาษาอังกฤษ
{name_en}

# สาขาการวิจัยหลัก
{oecd1}

# สาขาการวิจัยย่อย
{oecd2}

# คำสำคัญ
{keyword_th}

# คำสำคัญภาษาอังกฤษ
{keyword_en}

# วัตถุประสงค์
{objective}

# บทคัดย่อ
{abstract_th}

# บทคัดย่อภาษาอังกฤษ
{abstract_en}"""
    return text
```

```python
texts = []
for index, row in df.iterrows():
    texts.append(get_text(
        name_th=row["ProjectNameTH"],
        name_en=row["ProjectNameEN"],
        keyword_th=row["ProjectKeyword"],
        keyword_en=row["ProjectKeywordEng"],
        objective=row["ProjectObjective"],
        abstract_th=row["ProjectAbstract"],
        abstract_en=row["ProjectAbstractEN"],
        oecd1=row["OECD1"],
        
    ))
```
