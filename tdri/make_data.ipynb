```python
import pandas as pd
import os
import glob
```

```python
list_md = list(glob.glob(os.path.join(".","TDRI_MS-Word_Final-report","*.md")))
```

```python
list_md[0]
```

```python
list_data = []
for i in list_md:
   try:
      with open(i,"r") as f:
         list_data.append(f.read())
   except:
      continue
```

```python
len(list_data)
```

```python
list_data[0]
```

```python
import re
def remove_img_tags(data):
    p = re.compile(r'<img.*?/>')
    return p.sub('', data)
```

```python
df=pd.DataFrame.from_dict({
    "id":["tdri_"+str(i) for i in range(len(list_data))],
    "text":[remove_img_tags(i) for i in list_data],
    "source":"TDRI",
    "metadata":[{"domain":"literature"} for i in range(len(list_data))]
})
```

```python
df
```

```python
with open("tdri.jsonl","w") as f:
    f.write(df.to_json(orient='records', lines=True, force_ascii=False))
```
