```python
import pandas as pd
import glob
```

```python
list_jsonl=[i for i in list(glob.glob("*/*.jsonl")) if 'wiki' not in i] 
```

```python
list_jsonl
```

```python
import json
```

```python
data=[]
for i in list_jsonl:
    with open(i,"r") as f:
        for j in f:
            data.append(json.loads(j))
```

```python
data[0]
```

```python
len(data)
```

```python
with open("non-web-nowiki.jsonl","w") as f:
    for i in data:
        f.write(json.dumps(i,ensure_ascii=False) + "\n")
```
