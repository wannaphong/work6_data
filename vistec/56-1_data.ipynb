```python
import pandas as pd
```

```python
df=pd.read_csv("56-1_data.csv")
```

```python
df
```

```python
def make_json(id,text:str,source:str,metadata:dict=None):
    _temp = {
        "id":id,
        "text":str(text).strip(),
        "source":source
    }
    if metadata != None:
        _temp["metadata"]=metadata
    return _temp
```

```python
data=[]
for i,d in df.iterrows():
    data.append(make_json("secreport_"+str(i),d["text"].strip(),"airesearch/CMDF_VISTEC",{"ticker":d["Ticker"],"year":d["Year"],"Filename":d["Filename"],"domain":"finance"}))
```

```python
data[0]
```

```python
import json
with open("secreport.jsonl","w") as f:
    for i in data:
        f.write(json.dumps(i,ensure_ascii=False) + "\n")
```
