```python
from datasets import load_dataset

df = load_dataset("pythainlp/thai-constitution-corpus",split="train").to_pandas()
```

```python
def make_json(id,text:str,source:str,metadata:dict=None):
    _temp = {
        "id":id,
        "text":str(text).strip(),
        "source":source
    }
    if metadata != None:
        _temp["metadata"]=metadata
    return _temp
```

```python
df
```

```python
data=[]
for i,d in df.iterrows():
    data.append(make_json("thaiconstitution_"+str(i),d["txt"].strip(),"pythainlp/thai-constitution-corpus",{"title":d["name"],"domain":"legal"}))
```

```python
import json
```

```python
with open("thai-constitution-corpus.jsonl","w") as f:
    for i in data:
        f.write(json.dumps(i,ensure_ascii=False) + "\n")
```
