```python
from datasets import load_dataset

df = load_dataset("pythainlp/tlcv2.0_oa",split="train").to_pandas()
```

```python
def make_json(id,text:str,source:str,metadata:dict=None):
    _temp = {
        "id":id,
        "text":str(text).strip(),
        "source":source
    }
    if metadata != None:
        _temp["metadata"]=metadata
    return _temp
```

```python
df
```

```python
data=[]
for i,d in df.iterrows():
    data.append(make_json("tlc_"+str(i),d["TEXT"].strip(),"pythainlp/tlcv2.0_oa",{"domain":"literature"}))
```

```python
import json
```

```python
with open("tlcv2.0_oa.jsonl","w") as f:
    for i in data:
        f.write(json.dumps(i,ensure_ascii=False) + "\n")
```
