```python
import pandas as pd

df = pd.read_parquet("hf://datasets/pythainlp/thai-tnhc2-books/data/train-00000-of-00001.parquet")
```

```python
def make_json(id,text:str,source:str,metadata:dict=None):
    _temp = {
        "id":id,
        "text":str(text).strip(),
        "source":source
    }
    if metadata != None:
        _temp["metadata"]=metadata
    return _temp
```

```python
df
```

```python
data=[]
for i,d in df.iterrows():
    data.append(make_json("tnhc2_"+str(i),d["book"]+"\n\n"+d["text"].strip(),"pythainlp/thai-tnhc2-books",{"title":d["book"],"author":d["author"],"domain":"book"}))
```

```python
data[0]
```

```python
import json
```

```python
with open("thai-tnhc2-books.jsonl","w") as f:
    for i in data:
        f.write(json.dumps(i,ensure_ascii=False) + "\n")
```
