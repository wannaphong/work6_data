```python
from datasets import load_dataset

df = load_dataset("pythainlp/ratchagitja_v1.0",split="train").to_pandas()
```

```python
def make_json(id,text:str,source:str,metadata:dict=None):
    _temp = {
        "id":id,
        "text":str(text).strip(),
        "source":source
    }
    if metadata != None:
        _temp["metadata"]=metadata
    return _temp
```

```python
df2=df[df.apply(lambda x:  "ล้มละลาย" not in x['title'], axis=1)].reset_index()
```

```python
df2
```

```python
data=[]
for i,d in df2.iterrows():
    data.append(make_json("ratchagitja_"+str(i),d["text"].strip(),"pythainlp/ratchagitja_v1.0",{"domain":"legal","title":d["title"].strip(),"book":d["book"].strip(),"section":d["section"].strip(),"publication_date":d["publication_date"].strip(),"url":d["src"].strip()}))
```

```python
data[0]
```

```python
import json
```

```python
with open("ratchagitja_v1.0.jsonl","w") as f:
    for i in data:
        f.write(json.dumps(i,ensure_ascii=False) + "\n")
```
