```python
import pandas as pd
df=pd.read_json("Training data - academic literature TNRR.jsonl.old",lines=True).rename(columns={"uniq_id": "id","texts":"text"})
```

```python
df
```

```python
new_metdata=[]
_t = {"domain":"literature"}
for i in df["metadata"].to_list():
    del i["subset"]
    new_metdata.append({**_t,**i})
```

```python
new_metdata[0]
```

```python
df["metadata"]=new_metdata
```

```python
with open("tnrr.jsonl","w") as f:
    f.write(df.to_json(orient='records', lines=True, force_ascii=False))
```
